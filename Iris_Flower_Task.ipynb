#  Building a machine learning model that can learn from the measurements of the iris species and classify them

Load the dataset from sklearn library.

```python
from sklearn import datasets
iris=datasets.load_iris()
```

```python
print("\n Feature Names:",iris.feature_names,"\n Iris Data:",iris.data,"\n TargetNames:",iris.target_names,"\n Target:",iris.target)
```

Assign the data and target to separate variables

```python
x=iris.data
y=iris.target
```

```python
x
```

```python
y
```

-----*Splitting the dataset*-----      ||
x_train contains the training features ||
x_test contains the testing features   ||
y_train contains the training labels   ||
y_test contains the testing labels.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.5)
```

we can solve the problem by using any algorithm.
in this case we use "decision tree algorithm"

```python
from sklearn import tree
classifier=tree.DecisionTreeClassifier()
```

we can also use KNeighborsClassifier(efficiency is higher)

```python
from sklearn import neighbors
classifier=neighbors.KNeighborsClassifier()
```

```python
classifier.fit(x_train,y_train)
```

```python
predictions=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score: "+ str( accuracy_score(y_test,predictions)*100))
```

```python
diff=predictions-y_test
print("Result is: ")
print(diff)
```

```python
print('Total no of samples misclassifed =', sum(abs(diff)))
```
